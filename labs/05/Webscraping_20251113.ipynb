```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://www.scrapethissite.com/pages/forms/"
all_rows = []
row_id = 1

page = 1
while True:
    print(f"Scraping Seite {page} ...")
    # URL mit Parameter zusammensetzen
    url = f"{base_url}?page_num={page}"
    response = requests.get(url)
    
    if response.status_code != 200:
        print(f"Fehler beim Abruf der Seite {page}, Status: {response.status_code}")
        break

    soup = BeautifulSoup(response.text, "html.parser")

    # Prüfe, ob Tabelle auf der Seite ist
    table = soup.find("table")
    if not table:
        print(f"Keine Tabelle mehr auf Seite {page}, Abbruch.")
        break

    rows = table.find_all("tr")
    if not rows:
        print(f"Keine Datenzeilen gefunden auf Seite {page}, Abbruch.")
        break

    # Alle Zeilen auslesen und speichern
    for row in rows:
        #cells = row.find_all(["td", "th"])
        cells = row.find_all(["td"])
        cell_texts = [row_id] + [cell.get_text(strip=True) for cell in cells]
        all_rows.append(cell_texts)
        row_id += 1

    page += 1  # nächste Seite

    if page > 25:
        print ("Endlosschleife")
        break

# Falls Daten gesammelt wurden, erste Zeile als Header anpassen
if all_rows:
    all_rows[0][0] = "ID"

# In CSV schreiben
output_file = "output.csv"
with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    for row in all_rows:
        writer.writerow(row)

print(f"Scraping komplett. {row_id-1} Zeilen in '{output_file}' gespeichert.")
```

```python
import pandas as pd
df = pd.read_csv('output.csv')

df["Year"] = pd.to_numeric(arg=df["Year"])
df["Wins"] = pd.to_numeric(arg=df["Wins"])

jahre = [1990, 2000, 2010]
for jahr in jahre:
    df_jahr = df[df["Year"] == jahr]
    winmax = df_jahr["Wins"].idmax()
    top_team = df.loc[winmax]
    print(top_team)

jahre = [1991, 2001, 2011]
for jahr in jahre:
    df_jahr = df[df["Year"] == jahr]
    winmax = df_jahr["Wins"].idmax()
    top_team = df.loc[winmax]
    print(top_team)
```
